**Dataset_4**

```python
url_ds_4 = 'https://raw.githubusercontent.com/mflazaro/bases_ds_python/main/a_Dataset_CreditScoring.csv'

df_st_4 = pd.read_csv(url_ds_4, sep=';')

df_st_4
```

Caracteristicas Dataset_4:

TARGET: Target variable, where 0 indicates good credit and 1 indicates bad credit. /
ID: Customer ID or unique identifier. /
DerogCnt: Number of derogatory remarks on the customer's credit report. /
CollectCnt: Number of collections on the customer's credit report. /
BanruptcyInd: Indicates whether the customer has declared bankruptcy (0 for no, 1 for yes). /
InqCnt06: Number of credit inquiries made by the customer in the last 6 months. /
InqTimeLast: Time (in months) since the last credit inquiry. /
InqFinanceCnt24: Number of credit inquiries related to financing in the last 24 months. /
TLTimeFirst: Time (in months) since the first credit line was opened. /
TLTimeLast: Time (in months) since the last credit line was opened. /
TLBalHCPct: Percentage of total credit line balance to high credit/limit. /
TLSatPct: Percentage of accounts where the customer's credit balance is close to the credit limit. /
TLDel3060Cnt24: Number of accounts with late payments (30-60 days) in the last 24 months. /
TLDel90Cnt24: Number of accounts with late payments (90+ days) in the last 24 months. /
TLDel60CntAll: Total number of accounts with late payments (60+ days). /
TLOpenPct: Percentage of open accounts. /
TLBadDerogCnt: Number of accounts with bad derogatory remarks. /
TLDel60Cnt24: Number of accounts with late payments (60+ days) in the last 24 months. /
TLOpen24Pct: Percentage of open accounts in the last 24 months.
